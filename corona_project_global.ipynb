```python
def show_optional_counrties(user_input):
    optional_countries = []
    for country in data_infected_transpose.columns:
        input_length = len(user_input)
        if user_input[0:input_length] == country[0:input_length]:
            optional_countries.append(country)
    if optional_countries != []:
        print(optional_countries)
    else:
        print("No match was found")
countries_list = data_infected_transpose.columns
chosen_country = input('Choose a country: ')
while chosen_country not in countries_list:
    show_optional_counrties(chosen_country)
    chosen_country = input('Choose a country: ')   
```
